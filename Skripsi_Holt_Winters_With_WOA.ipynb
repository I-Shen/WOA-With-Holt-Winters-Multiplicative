# **FIX HOLT-WINTERS**

```python
def holtwinters (dt,maxPop,alpha,beta,gamma,seasLenght,predLenght) :
  print("Parameter Holt Winter Sekarang :")
  for y in range(0,maxPop) :
    print(alpha[y])
    print(beta[y])
    print(gamma[y])
    print()
    y = y + 1
  #create rata2 value
  rata2 = stat.mean(dt['Jumlah'])
  #print(rata2)
  extInd = len(dt['Jumlah'])+predLenght
  #print(tambah)
  #for in masing2 populasi
  mape = [None] * maxPop
  for arf in range(0, maxPop) :
  #print("POPULASI HW KE-" + str(arf) + "-------------------------------------------------")
  #initial list
    season = [None] * extInd
    tren = [None] * extInd
    level = [None] * extInd
    predik = [None] * extInd
    error = [None] * extInd
    ape = [None] * len(dt['Jumlah'])
    #inape itu sebagai bantu index ape dimulai dari index 0
    inape = 0
    #k itu sebagai bantu index untuk variabel season yang berbeda urutan index dengan variabel lain dimulai dari index 1
    k = 0
    #for in season awal(level, tren, season, predik, error)
    for fsawal in range(0, seasLenght) :
      season[fsawal] = dt['Jumlah'][fsawal]/rata2
      tren[fsawal] = 0
      level[fsawal] = 0
      predik[fsawal] = 0
      error[fsawal] = 0
      if fsawal == seasLenght :
        tren[fsawal] = 0
        level[fsawal] = dt['Jumlah'][fsawal]/season[fsawal]
      fsawal = fsawal + 1
    for fslanjut in range(fsawal,(len(dt['Jumlah']))) :
      level[fslanjut] = alpha[arf]*(dt['Jumlah'][fslanjut])/season[k]+(1-alpha[arf])*(level[fslanjut-1]+tren[fslanjut-1])
      tren[fslanjut] = beta[arf]*(level[fslanjut]-level[fslanjut-1])+(1-beta[arf])*tren[fslanjut-1]
      season[fslanjut] = gamma[arf]*(dt['Jumlah'][fslanjut]/level[fslanjut])+(1-gamma[arf])*season[k]
      predik[fslanjut] = (level[fslanjut-1]+tren[fslanjut-1])*season[k]
      error[fslanjut] = dt['Jumlah'][fslanjut]-predik[fslanjut]
      if dt['Jumlah'][fslanjut] == 0 :
        ape[inape] = 0
      else :
        ape[inape] = (abs(error[fslanjut])/dt['Jumlah'][fslanjut])*100
      k = k + 1
      inape = inape + 1
      fslanjut = fslanjut + 1
    # #remove None value in list(level, tren, error, ape)
    level = [h1 for h1 in level if h1 is not None]
    tren = [h2 for h2 in tren if h2 is not None]
    error = [h3 for h3 in error if h3 is not None]
    ape = [h4 for h4 in ape if h4 is not None]
    #for in prediksi(season & predik)
    for fsakhir in range(fslanjut, fslanjut+predLenght) :
      season[fsakhir] = gamma[arf]*(dt['Jumlah'][fslanjut-1]/level[fslanjut-1])+(1-gamma[arf])*season[k]
      predik[fsakhir] = (level[fslanjut-1]+tren[fslanjut-1])*season[k]
      k = k + 1
    # #process mape in one population
    mape[arf] = np.mean(ape)
    print("MAPE Pada Populasi Ke-" + str(arf+1) + " = " + str(mape[arf]) + " ---------------------")
    print()
    arf = arf + 1
  #print("Hasil Algoritma Holt-Winters Pada Iterasi Ke-" + str(t) + " = -------------------------------------")
  #create minimal mape
  minMape = min(mape)
  # print(minMape)
  #create maximal mape
  maxMape = mape[0]
  max_val_idx = 0
  for innn in range(len(mape)):
    if mape[innn] > maxMape:
      maxMape = mape[innn]
      max_val_idx = innn
    innn = innn + 1
  #check minimal & maximal value
  print("Minimal Mape " + " = " + str(minMape))
  print("Maksimal Mape " + " = " + str(maxMape))
  #get index minimal mape in lisf of mape, array start from 0
  minInMape = mape.index(minMape)
  maxInMape = mape.index(maxMape)
  #print(minInMape)
  print("Index Minimal Mape Iterasi Ke-" + " = " + str(minInMape+1))
  print("Index Maksimal Mape Iterasi Ke-" + " = " + str(maxInMape+1))
  print()
  #get great position of population in this iteration, array start from 0
  print("Alpha = " + str(alpha[minInMape]))
  print("Beta = " + str(beta[minInMape]))
  print("Gamma = " + str(gamma[minInMape]))
  print("Alpha = " + str(alpha))
  print("Beta = " + str(beta))
  print("Gamma = " + str(gamma))
  print("----------------------------------------------------------------------")
  return  minMape, minInMape, alpha[minInMape], beta[minInMape], gamma[minInMape], fslanjut, predik
```

# **FIX WHALE OPTIMIZATION ALGORITHM AND HYBRID**

```python
#import package
import pandas as pd
import random
from random import choice
import numpy as np
import statistics as stat
import math
import matplotlib.pyplot as plt

#import dataset
dt = pd.read_excel("ABC Kecap Asin 133ML.xlsx")
#dt = pd.read_excel("B29 Pink Softener.xlsx")

#input constanta
maxIters = input("Maksimal Iterasi = ")
maxPops = input("Maksimal Populasi = ")
sLenghts = input("Jumlah Panjang Seasonal = ")
predz = input("Jumlah Periode Prediksi = ")
maxPop = int(maxPops)
seasLenght = int(sLenghts)
predLenght = int(predz)
maxIter = int(maxIters)
#check for important constanta
print("Maksimal Iterasi Yang Dimasukkan = " + str(maxIter))
print("Maksimal Populasi Yang Dimasukkan = " + str(maxPop))
print("Jumlah Panjang Seasonal Yang Dimasukkan = " + str(seasLenght))
print("Jumlah Periode Prediksi Yang Dimasukkan = " + str(predLenght))

#declaration param
t = 0
blt = 7
bestMapeSaatIni = 0
bestAlphaSaatIni = 0
bestBetaSaatIni = 0
bestGammaSaatIni = 0
kurangi = 0
#arrmaxIter is maximum iteration but cz iteration start from 1, we should +1 to sync array lenght
arrmaxIter = maxIter + 1
#declaration array lenght
alpha = [None] * maxPop
beta = [None] * maxPop
gamma = [None] * maxPop
alphaTemp = [None] * maxPop
betaTemp = [None] * maxPop
gammaTemp = [None] * maxPop
bestPrediksiSaatIni = [None] *(len(dt['Jumlah']))
mapeIter = [None] * maxIter

#first initial location
for j in range(0,maxPop):
  alpha[j] = random.uniform(0,1)
  beta[j] = random.uniform(0,1)
  gamma[j] = random.uniform(0,1)
  alpha[j] = round(alpha[j],blt)
  beta[j] = round(beta[j],blt)
  gamma[j] = round(gamma[j],blt)
  j = j + 1

print("bestMapeSaatIni = " + str(bestMapeSaatIni))
print("bestAlphaSaatIni = " + str(bestAlphaSaatIni))
print("bestBetaSaatIni = " + str(bestBetaSaatIni))
print("bestGammaSaatIni = " + str(bestGammaSaatIni))
print("bestPrediksiSaatIni[-kurangi:] = " + str(bestPrediksiSaatIni[-kurangi:]))
print("bestPrediksiSaatIni = " + str(bestPrediksiSaatIni))
print("List Alpha = " + str(alpha))
print("List Beta = " + str(beta))
print("List Gamma = " + str(gamma))
print("List Temp Alpha = " + str(alphaTemp))
print("List Temp Beta = " + str(betaTemp))
print("List Temp Gamma = " + str(gammaTemp))

minMape, minInMape, alpha[minInMape], beta[minInMape], gamma[minInMape], fslanjut, predik = holtwinters(dt,maxPop,alpha,beta,gamma,seasLenght,predLenght)
print(alpha)
print(beta)
print(gamma)
#take some variable for result of documentation
#var temp for best param
bestMapeSaatIni = minMape
bestAlphaSaatIni = alpha[minInMape]
bestBetaSaatIni = beta[minInMape]
bestGammaSaatIni = gamma[minInMape]
bestPrediksiSaatIni = predik
mapeIter[0] = bestMapeSaatIni
kurangi = (fslanjut+predLenght) - fslanjut
while t < maxIter:
  posisi1 = [None] * maxPop
  posisi2 = [None] * maxPop
  posisi3 = [None] * maxPop  
  print("--------------------------------------------ITERASI KE-" + str(t+1) + "-------------------------------------------------")
  print("bestMapeSaatIni = " + str(bestMapeSaatIni))
  print("bestAlphaSaatIni = " + str(bestAlphaSaatIni))
  print("bestBetaSaatIni = " + str(bestBetaSaatIni))
  print("bestGammaSaatIni = " + str(bestGammaSaatIni))
  print("bestPrediksiSaatIni[-kurangi:] = " + str(bestPrediksiSaatIni[-kurangi:]))
  print("bestPrediksiSaatIni = " + str(bestPrediksiSaatIni))
  print("List Alpha = " + str(alpha))
  print("List Beta = " + str(beta))
  print("List Gamma = " + str(gamma))
  print("List Temp Alpha = " + str(alphaTemp))
  print("List Temp Beta = " + str(betaTemp))
  print("List Temp Gamma = " + str(gammaTemp))
  a = 2-t*((2)/maxIter)
  a2 = (-1)+t*((-1)/maxIter)
  #lets process location of each population
  for i in range(0,maxPop):
    r1 = random.uniform(0,1)
    r2 = random.uniform(0,1)
    A = 2*a*r1-a
    C = 2*r2
    l = (a2-1)*random.uniform(0,1)+1
    b = 1
    p = random.uniform(0,1)
    #check vector value
    print()
    print("Berikut Merupakan Isi Dari Variabel Vector yang Akan Kita Proses Untuk Iterasi Ke-" + str(t+1) + " = ")
    print("r1 = " + str(r1))
    print("r2 = " + str(r2))
    print("a = " + str(a)) 
    print("a2 = " + str(a2))
    print("A = " + str(A))
    print("C = " + str(C))
    print("l = " + str(l))
    print("b = " + str(b))
    print("p = " + str(p))
    print()
   #decision for next location of pray was determined from whale pattern
    if p < 0.5:
      print("p < 0.5")
      if abs(A) < 1:
        print("masuk if abs(A) < 1")
        D1 = abs(C*bestAlphaSaatIni-alpha[i])
        D2 = abs(C*bestBetaSaatIni-beta[i])
        D3 = abs(C*bestGammaSaatIni-gamma[i])
        posisi1 = bestAlphaSaatIni-A*D1
        posisi2 = bestBetaSaatIni-A*D2
        posisi3 = bestGammaSaatIni-A*D3
        posisi1 = round(posisi1,blt)
        posisi2 = round(posisi2,blt)
        posisi3 = round(posisi3,blt)
        if ((posisi1>1) or (posisi2>1) or (posisi3>1)) or ((posisi1<0) or (posisi2<0) or (posisi3<0)) or ((posisi1==0) or (posisi2==0) or (posisi3==0)):
          posisi1 = bestAlphaSaatIni
          posisi2 = bestBetaSaatIni
          posisi3 = bestGammaSaatIni
      elif abs(A) >= 1:
        print("masuk if abs(A) >= 1")
        alphaR = alpha[round(random.uniform(0,maxPop-1))]
        betaR = beta[round(random.uniform(0,maxPop-1))]
        gammaR = gamma[round(random.uniform(0,maxPop-1))]
        D1 = abs((C*alphaR)-alpha[i])
        D2 = abs((C*betaR)-beta[i])
        D3 = abs((C*gammaR) - gamma[i])
        posisi1 = alphaR-A*D1
        posisi2 = betaR-A*D2
        posisi3 = gammaR-A*D3
        posisi1 = round(posisi1,blt)
        posisi2 = round(posisi2,blt)
        posisi3 = round(posisi3,blt)
        if ((posisi1>1) or (posisi2>1) or (posisi3>1)) or ((posisi1<0) or (posisi2<0) or (posisi3<0)) or ((posisi1==0) or (posisi2==0) or (posisi3==0)):
          posisi1 = bestAlphaSaatIni
          posisi2 = bestBetaSaatIni
          posisi3 = bestGammaSaatIni
    elif p >= 0.5:
      print("p >= 0.5")
      D1 = abs(bestAlphaSaatIni-alpha[i])
      D2 = abs(bestBetaSaatIni-beta[i])
      D3 = abs(bestGammaSaatIni-gamma[i])
      posisi1 = D1*math.exp(b*l)*math.cos(l*2*math.pi)+bestAlphaSaatIni
      posisi2 = D2*math.exp(b*l)*math.cos(l*2*math.pi)+bestBetaSaatIni
      posisi3 = D3*math.exp(b*l)*math.cos(l*2*math.pi)+bestGammaSaatIni
      posisi1 = round(posisi1,blt)
      posisi2 = round(posisi2,blt)
      posisi3 = round(posisi3,blt)
      if ((posisi1>1) or (posisi2>1) or (posisi3>1)) or ((posisi1<0) or (posisi2<0) or (posisi3<0)) or ((posisi1==0) or (posisi2==0) or (posisi3==0)):
        posisi1 = bestAlphaSaatIni
        posisi2 = bestBetaSaatIni
        posisi3 = bestGammaSaatIni
    alphaTemp[i] = posisi1
    betaTemp[i] = posisi2
    gammaTemp[i] = posisi3
  #   print()
    i = i + 1
  for bTemp in range(0,maxPop) :
    alpha[bTemp] = alphaTemp[bTemp]
    beta[bTemp] = betaTemp[bTemp]
    gamma[bTemp] = gammaTemp[bTemp]
    bTemp = bTemp + 1
  print()
  #---------------------------------------------------------------------
  print("bestMapeSaatIni = " + str(bestMapeSaatIni))
  print("bestAlphaSaatIni = " + str(bestAlphaSaatIni))
  print("bestBetaSaatIni = " + str(bestBetaSaatIni))
  print("bestGammaSaatIni = " + str(bestGammaSaatIni))
  print("bestPrediksiSaatIni[-kurangi:] = " + str(bestPrediksiSaatIni[-kurangi:]))
  print("bestPrediksiSaatIni = " + str(bestPrediksiSaatIni))
  print("List Alpha = " + str(alpha))
  print("List Beta = " + str(beta))
  print("List Gamma = " + str(gamma))
  print("List Temp Alpha = " + str(alphaTemp))
  print("List Temp Beta = " + str(betaTemp))
  print("List Temp Gamma = " + str(gammaTemp))
  #HOLT-WINTER ALGORITHM AGAIN------------------------------------------
  minMape, minInMape, alpha[minInMape], beta[minInMape], gamma[minInMape], fslanjut, predik = holtwinters(dt,maxPop,alpha,beta,gamma,seasLenght,predLenght)
  #update if last mape bigger than mape now in each iteration
  if minMape < bestMapeSaatIni :
    print("Mape lebih kecil dari sebelumnya ternyata, maka param nya tidak perlu diupdate")
    bestMapeSaatIni = minMape
    bestAlphaSaatIni = alpha[minInMape]
    bestBetaSaatIni = beta[minInMape]
    bestGammaSaatIni = gamma[minInMape]
    bestPredikiSaatIni = predik
  #if not, bestSolution param are not change. And using last value as param in next iteration
  else :
    print("Mape lebih besar dari sebelumnya ternyata, maka param nya tidak perlu diupdate")
    print()
  print()
  mapeIter[t] = bestMapeSaatIni
  #-----------------------------------------------------------------
  t = t + 1
print()
bestPrediksiSaatIni = list(map(round,bestPrediksiSaatIni))
#RESULT--------------------------------------------------------------
print("bestMapeSaatIni = " + str(bestMapeSaatIni))
print("bestAlphaSaatIni = " + str(bestAlphaSaatIni))
print("bestBetaSaatIni = " + str(bestBetaSaatIni))
print("bestGammaSaatIni = " + str(bestGammaSaatIni))
print("bestPrediksiSaatIni[-kurangi:] = " + str(bestPrediksiSaatIni[-kurangi:]))
print("bestPrediksiSaatIni = " + str(bestPrediksiSaatIni))
print("List Alpha = " + str(alpha))
print("List Beta = " + str(beta))
print("List Gamma = " + str(gamma))
print("List Temp Alpha = " + str(alphaTemp))
print("List Temp Beta = " + str(betaTemp))
print("List Temp Gamma = " + str(gammaTemp))
print()
print("MAPE semua ITER = ")
for rekap in range(0, maxIter):
  print(mapeIter[rekap])
#Plot MAPE per Iteration
import matplotlib.pyplot as plt
x_axis = [i for i in range(maxIter)]
y_axis = mapeIter 
plt.plot(x_axis, y_axis)
plt.title('Plot MAPE per Iterasi')
plt.xlabel('Iterasi')
plt.ylabel('MAPE')
plt.show()
```
